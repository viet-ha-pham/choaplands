# Các hàm tiện ích

```python
def trans(txt):
    sp = txt.split()
    return " ".join([ vie_cho_dct[w] if w in vie_cho_dct else w for w in sp]).replace("lồn","noòng").replace("khỏe","tãy")
```

```python
def normalize_word(word):
    word = word.replace(".","")
    word = word.replace(",","")
    word = word.replace("?","")
    word = word.replace("!","")
    word = word.replace(":","")
    word = word.replace("(","")
    word = word.replace(")","")
    word = word.replace("\"","")

    return word
```

```python
def make_vocab(words):
    l1 = [trans(normalize_word(w)) for w in words]
    l2 = [normalize_word(w) for w in words]
    words = []
    assert len(l1) == len(l2)
    
    for i in range(len(l1)):
        w = f"{l1[i]}: {l2[i]}"
        words.append(w)
    res = "\n".join(list(set(words)))
    res = f"Từ vựng/{trans('từ vựng')}\n\n" + res
    return res
```

```python
def beautify(s):
    # viết hoa chữ đầu tiên và chỉnh lại vị trí dấu chấm phẩy v.v...
    res = s[0].upper() + s[1:]
    res = res.replace(" .",".")
    res = res.replace(" ,",",")
    res = res.replace(" ?","?")
    res = res.replace(" !","!")
    res = res.replace(" (","(")
    res = res.replace(" )",")")
    res = res.replace(" \"","\"")
    res = res.replace(" :",":")
    return res
```

```python
def normalize_sentence(s):
    # viết thường chữ đầu và chỉnh lại vị trí dấu chấm phẩy v.v...
    res = s[0].lower() + s[1:]
    res = res.replace("."," .")
    res = res.replace(","," ,")
    res = res.replace("?"," ?")
    res = res.replace("!"," !")
    res = res.replace("("," (")
    res = res.replace(")"," )")
    res = res.replace("\""," \"")
    res = res.replace(":"," :")
    return res
```

```python
def make_examples(sentences):
    l1 = [ trans(normalize_sentence(s)) for s in sentences ]
    l2 = sentences
    pairs = []
    assert len(l1) == len(l2)
    
    for i in range(len(l1)):
        p = f"{beautify(l1[i])}\n{beautify(l2[i])}"
        pairs.append(p)
    res = "\n\n".join(pairs)
    res = "Ví dụ/Nghỉn nhoảo\n\n" + res
    return res
```

```python
def make_unit(title, sentences):
    words = []
    for s in sentences:
        words.extend([ w for w in s.split() if w[0].upper()!=w[0] and w not in ".,?!():\""])
    words = list(set(words))

    return f"""{beautify(title).upper()}/{beautify(trans(title)).upper()}
    
    
{make_vocab(words)}
    
    
{make_examples(sentences)}

#choap
    """
```

```python
def make_unit_from_txt(file):
    with open(file, "r", encoding="utf-8") as f:
        lines = f.readlines()
    lines = [ beautify(l.strip()) for l in lines if l.strip() ]
    title = normalize_sentence(lines[0])
    sentences = lines[1:]
    words = []
    for s in sentences:
        words.extend([ w for w in s.split() if w[0].upper()!=w[0] and w not in ".,?!():\""])
    words = list(set(words))

    return f"""{beautify(title).upper()}/{beautify(trans(title)).upper()}
    
    
{make_vocab(words)}
    
    
{make_examples(sentences)}

#choap
    """
```

```python
import os
import glob

def tim_file_moi_nhat(duong_dan):
    danh_sach_files = glob.glob(os.path.join(duong_dan, '*'))
    if not danh_sach_files:
        return None
    file_moi_nhat = max(danh_sach_files, key=os.path.getctime)
    return file_moi_nhat
```

```python
def make_unit_from_folder(path):
    file_moi_nhat = tim_file_moi_nhat(path)
    if file_moi_nhat:
        print(make_unit_from_txt(file_moi_nhat))
    else:
        print("Không tìm thấy file trong thư mục.")
```

```python
import random

def write_poem(keyword1, keyword2, keyword3):
    # Tạo danh sách các câu thơ có thể
    phrases = [
        f"Tôi nhìn {keyword1} và",
        f"Một chiếc {keyword2} đầy {keyword3}",
        f"Biết bao {keyword1} trên đường",
        f"{keyword2} và {keyword3}",
        f"Rung động lòng những {keyword1} kia",
        f"Như những {keyword2} ngọt ngào",
        f"{keyword3} bay lượn trong gió",
        f"Và {keyword1} tràn ngập không gian",
        f"Khi {keyword2} và {keyword3} hòa quyện",
        f"Trong những đêm trăng {keyword1}"]
    
    # Lựa chọn ngẫu nhiên các câu thơ để tạo thành bài thơ
    poem = random.sample(phrases, k=4)
    
    # Trả về bài thơ đã được tạo
    return "\n".join(poem)

# Sử dụng hàm để viết bài thơ
keyword1 = "hoa"
keyword2 = "mây"
keyword3 = "mùa xuân"
poem = write_poem(keyword1, keyword2, keyword3)
print(poem)
```

```python
import random

def write_poem(keywords, num_lines=4):
    phrases = []
    poem = []

    # Tạo danh sách các câu thơ từ các từ khóa
    for keyword in keywords:
        phrases.append(f"{keyword} rực rỡ")
        phrases.append(f"Một {keyword} tuyệt đẹp")
        phrases.append(f"Như {keyword} trong mơ")
        phrases.append(f"Khắp nơi {keyword} tỏa sáng")

    # Lựa chọn ngẫu nhiên các câu thơ
    selected_phrases = random.sample(phrases, num_lines)

    # Xây dựng bài thơ từ các câu thơ đã chọn
    for i in range(num_lines):
        line = selected_phrases[i]
        if i % 2 == 0:
            line += ","
        else:
            line += "."
        poem.append(line)

    # Trả về bài thơ đã được tạo
    return "\n".join(poem)

# Sử dụng hàm để viết bài thơ
keywords = ["hoa", "mây", "mùa xuân", "gió"]
poem = write_poem(keywords, num_lines=6)
print(poem)
```

```python
def print_triangular_text(text):
    words = text.split()[:28]  # Chỉ lấy 28 từ đầu tiên
    num_words = len(words)
    line_length = 1  # Độ dài của dòng bắt đầu từ 1

    for i in range(num_words):
        print(' '.join(words[:line_length]))
        words = words[line_length:]
        line_length += 1
        if line_length > 7:  # Nếu đạt độ dài dòng là 7, đặt lại độ dài dòng thành 1
            line_length = 1

# Sử dụng hàm với một đoạn văn bản bất kỳ
text = """Trăm năm trong cõi người ta 
Chữ tài chữ mệnh khéo là ghét nhau
Trải qua một cuộc bể dâu
Những điều trông thấy mà đau đớn lòng
"""
print_triangular_text(text)
```
